```python
#8 . Explore over an internet that how much calories burn is required for a healthy person and find out how many healthy person we have in our dataset

#For Women - 2000
#For men - 2500

#I have considered 2000 as the threshold

#I have calculated the average calorie burnt by person & compared it with the threshold to find if they are healthy or not.
```

```python
avg_calorie = fitbit_df[["Calories","Id"]].groupby("Id").mean().reset_index()
```

```python
calorie_df = pd.DataFrame(avg_calorie)
```

```python
def cal_burn(c):
  if c > 2000:
    return "Healthy"

  else:
    return "Unhealthy"

calorie_df["Health_Status"] = calorie_df["Calories"].apply(cal_burn)
calorie_df
```
